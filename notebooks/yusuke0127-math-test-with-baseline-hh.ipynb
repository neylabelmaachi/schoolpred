## Data prep for test results

```python
test_results = pd.read_csv("Morocco_CCT_Education_Child_Math_Test_Results.csv")
test_results.head()
```

```python
rename_col_test_results = {
    # Test
    "t1": "done_test",
    "t3_1": "digit_recognition_res",
    "t4_1": "number_recognition_res",
    "t5_1": "subtraction_res",
    "t6": "division_res",
}
```

```python
test_results = test_results.rename(columns=rename_col_test_results)
test_results.head()
```

```python
test_results.columns
```

```python
col_to_drop = ["year","round","endline", "studycode","unitobs","countrycode","ruralcode", "gendercode", "agecode", 't3_2', 't4_2', 't5_rep1', 't5_rep2', 't5_2', 't6_rep', 'hhid_endline', "surveycode", "region", "province", "t2_1", "t2_2", "prenom_enf_test", "schoolid"]
```

```python
test_results = test_results.drop(columns=col_to_drop)
test_results.head()
```

```python
test_results.shape
```

## Combining baseline household and test results

```python
baseline_with_test_results = pd.merge(filtered_baseline, test_results, on="hhid")
baseline_with_test_results.head()
```
